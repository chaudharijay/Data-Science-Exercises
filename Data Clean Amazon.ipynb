```python
df = df.drop(columns = ['Unnamed: 0', 'asin', 'product_url', 'product_photo',
       'product_num_offers', 'product_minimum_offer_price','is_amazon_choice', 'is_prime', 'climate_pledge_friendly',
       'sales_volume', 'delivery', 'has_variations'])
```

```python
df
```

```python
df_new_column_names = []

for i in df.columns:
    df_new_column_names.append(i.capitalize())

df_new_column_names
```

```python
df.columns  = df_new_column_names
df
```

```python
data = df
```

```python
data.duplicated().sum()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data = data[["Product_title", "Product_price", "Product_star_rating", "Product_num_ratings", "Is_best_seller"]]
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum().sum()
```

```python
data["Is_best_seller"].unique()
data["Is_best_seller"] = data["Is_best_seller"].apply(lambda x:1 if x == True else 0)
data
```

```python
data["Product_price"] = data["Product_price"].str.replace("$", "")
data["[Product_price"] = data["Product_price"].astype(float)
```
